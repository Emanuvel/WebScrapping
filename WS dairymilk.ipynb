## website

```python
url='https://www.flipkart.com/search?q=chocolate&sid=eat%2C0pt&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_3_na_na_na&as-pos=2&as-type=HISTORY&suggestionId=chocolate%7CChocolates&requestId=8496499f-afb7-4377-bdfb-ac27f13dfc05'
```

## website response

```python
res = requests.get(url)
```

```python
res
```

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

## dairy milk chocolates

```python
chocoltes_names=soup.select('.s1Q9rs')
```

```python
chocoltes_names
```

```python
chocolates_prices=soup.select('._8VNy32 ._30jeq3')
```

```python
chocolates_prices
```

```python
chocolates_offers=soup.select('._8VNy32 ._3Ay6Sb')
```

```python
chocolates_offers
```

```python
df=silk.DataFrame(chocoltes_names,columns=['chocoltes_names'])
```

```python
df
```

```python
df['chocolates_prices']=chocolates_prices
```

```python
df['chocolates_offers']=chocolates_offers
```

```python
df
```

```python
df.to_csv('dairymilk.csv')
```
